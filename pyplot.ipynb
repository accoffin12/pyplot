# Question 1:

Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).


### Answer Overiview:
There are actually two methods that can be used to answer this question. One that utilizes pandas and the other a quick overview. There are advantages and disadvantages to each. 
#### Pandas DataFrames:
With the first method utilizing pandas, the text is transfered into a DataFrame that can be referenced later and utilized for a larger amount of analysis. While in this example it isn't as critical, when dealing with machine learning or heavy statistical analysis transfering data into a dataframe allows for a larger amount of manipulation. For a majority of these questions I will be utilizing this method.

#### Quick Method:
The quick method that's demonstrated in the second box allows us to obtain an over view of the data, and have a little more control in terms of how the plot is created. However, the string is still a string and each time that we go to work with the information we will have to remember the string formatting which can be limiting. On the bright side this method allows for us to quickly obtain a view of a graph. 

```python
""" _Pandas Example_"""
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
TEXT = "At three o’clock precisely I was at Baker Street, but Holmes had not yet returned. The landlady informed me that he had left the house shortly after eight o’clock in the morning. I sat down beside the fire, however, with the intention of awaiting him, however long he might be. I was already deeply interested in his inquiry, for, though it was surrounded by none of the grim and strange features which were associated with the two crimes which I have already recorded, still, the nature of the case and the exalted station of his client gave it a character of its own. Indeed, apart from the nature of the investigation which my friend had on hand, there was something in his masterly grasp of a situation, and his keen, incisive reasoning, which made it a pleasure to me to study his system of work, and to follow the quick, subtle methods by which he disentangled the most inextricable mysteries. So accustomed was I to his invariable success that the very possibility of his failing had ceased to enter into my head.".lower()
letter_freq = dict(Counter(TEXT))
# Creates a printed string that allows for us to determine if the bar graph is correct.
# Please note this data will include ' ' , and . as we are counting characters. Since we didn't specify only letters I have lef these.
text_df = pd.DataFrame(list(letter_freq.items()), columns=['Character', 'Frequency'])
display(text_df)
ax = text_df.plot.bar(x='Character', y='Frequency')
ax.set_xlabel('Characters')
ax.set_ylabel('Frequency')
ax.set_title('Character Frequency')

plt.show()
```

# Question 2:
Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

### Answer Overview:
For this question there are multiple ways to generate random numbers. I decided to do the easiest method which is using numpy and then dictating the range that it could use as well as the size fo the sample.

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating 5 randome intigers between 0 and 25 using random from numpy.
random_numbers=np.random.randint(1, 25, size=5)

#Creating the scatter plot
fig1=plt.scatter(range(len(random_numbers)), random_numbers)
plt.title('Visual of Random Number Scatter Graph')
plt.figure(figsize=(9, 4))
print(fig1)

#Creating the graph using plot
fig2=plt.plot(range(len(random_numbers)), random_numbers)
plt.title('Visual of Random Number Plot')
plt.figure(figsize=(9, 4))
print(fig2)
```

# Question 3:
Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

#Generating 5 randome intigers between 1 and 50 using random from numpy.
random_numbers=np.random.randint(1, 50, size=5)

#Creating the scatter plot with different shape
fig3=plt.scatter(range(len(random_numbers)), random_numbers, marker='x', color='red')
plt.title('Visual of Scatter with Random Number Numpy')
plt.figure(figsize=(9, 4))
print(fig3)

#Creating a plot using a different line style
fig4=plt.plot(range(len(random_numbers)), random_numbers, linestyle='--', color='purple')
plt.title('Visual of Plot with Random Number Numpy')
plt.figure(figsize=(9, 4))
print(fig4)
```

# Question 4:
Given the data sets below, plot or scatter both on the same set of axes.  You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

### Answer Overview:
The seperate sets were transfered into a DataFrame which allowed for smaller and lighter code when creating the plot. This was ideal as it decreased the run time and allowed for us to easily plot both insertion_sort_times and merge_sort_times. 

```python
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

# Creating a plot using two features from a df
import matplotlib.pyplot as plt
import pandas as pd

#Creating the DataFrame
list_df = pd.DataFrame({'sizes': sizes, 'insertion_sort_times': insertion_sort_times, 'merge_sort_times': merge_sort_times})
display(list_df)

#Creating the plot
list_df.plot(x='sizes', y=['insertion_sort_times', 'merge_sort_times'])
plt.ylabel('Time (s)')
plt.show()
```
